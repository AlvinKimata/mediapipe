```python
import cv2
import math
import numpy as np
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = 'inputs/kimata.jpg'
```

```python
DESIRED_HEIGHT = 480
DESIRED_WIDTH = 480

def resize_and_show(image):
    h, w = image.shape[:2]
    if h < w:
        img = cv2.resize(image, (DESIRED_WIDTH, math.floor(h / (w / DESIRED_WIDTH))))
    else:
        img = cv2.resize(image, (math.floor(w / (h / DESIRED_HEIGHT)), DESIRED_HEIGHT))
    plt.imshow(img)
```

```python
IMAGE = plt.imread(IMAGE_PATH)
```

```python
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
```

```python
with mp_holistic.Holistic(min_detection_confidence=0.5, 
                          min_tracking_confidence=0.5) as holistic:
    
    results = holistic.process(cv2.cvtColor(IMAGE, cv2.COLOR_BGR2RGB))
    
    image_height, image_width, _ = IMAGE.shape
    
    #Draw pose landmarks.
    annotated_image = IMAGE.copy()
    
    #Right hand landmarks.
    mp_drawing.draw_landmarks(annotated_image, results.right_hand_landmarks, 
                              mp_holistic.HAND_CONNECTIONS)
    
    #Face landmarks.
    mp_drawing.draw_landmarks(annotated_image, results.face_landmarks, 
                              mp_holistic.FACEMESH_TESSELATION,
                              landmark_drawing_spec = None, 
                              connection_drawing_spec = mp_drawing_styles.get_default_face_mesh_tesselation_style())
    
    #Pose landmarks.
    mp_drawing.draw_landmarks(annotated_image, results.pose_landmarks,
                              mp_holistic.POSE_CONNECTIONS)
    
    resize_and_show(annotated_image)
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('No frames found.')
        break
        
    IMAGE = frame.copy()
    
    with mp_holistic.Holistic(min_detection_confidence=0.5, 
                          min_tracking_confidence=0.5) as holistic:
    
        results = holistic.process(cv2.cvtColor(IMAGE, cv2.COLOR_BGR2RGB))

        image_height, image_width, _ = IMAGE.shape

        #Draw pose landmarks.
        annotated_image = IMAGE.copy()

        #Right hand landmarks.
        mp_drawing.draw_landmarks(annotated_image, results.right_hand_landmarks, 
                                  mp_holistic.HAND_CONNECTIONS)
        
        #Left hand landmarks.
        mp_drawing.draw_landmarks(annotated_image, results.left_hand_landmarks, 
                                  mp_holistic.HAND_CONNECTIONS)
        
        #Face landmarks.
        mp_drawing.draw_landmarks(annotated_image, results.face_landmarks, 
                                  mp_holistic.FACEMESH_TESSELATION,
                                  landmark_drawing_spec = None, 
                                  connection_drawing_spec = mp_drawing_styles.get_default_face_mesh_tesselation_style())

        # #Pose landmarks.
        # mp_drawing.draw_landmarks(annotated_image, results.pose_landmarks,
        #                           mp_holistic.POSE_CONNECTIONS)
        
        cv2.imshow('image', annotated_image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
cap.release()
cv2.destroyAllWindows()
```
